```python
def safe_replace(value, old, new):
    return value.replace(old, new) if value else None

def url1(data,i,no,data_dict):
    for item in data_dict['response']['body']['items']['item']:
        if isinstance(item, dict) and 'applicationNumber' in item:
            if item['applicationNumber'] == no:
                item['title'] = data['data'][i]['title']
                item['entire_text'] = data['data'][i]['summary_entire'][0]['orginal_text']
                item['section_text'] = data['data'][i]['summary_section'][0]['orginal_text']
                return item
            else :
                return '없음'
        else :
            return '없음'
        
def url2(data,i,data_dict):
    item = data_dict['response']['body']['items']['item']
    if isinstance(item, dict) :
        item['title'] = data['data'][i]['title']
        item['entire_text'] = data['data'][i]['summary_entire'][0]['orginal_text']
        item['section_text'] = data['data'][i]['summary_section'][0]['orginal_text']
        return item
    else :
        return '없음'

def url3(data,i,no,data_dict):
    item = {}
    if isinstance(data_dict['response']['body']['item'], dict) :
        if isinstance(data_dict['response']['body']['item']['applicantInfoArray']['applicantInfo'], list):
            names_list = [info['name'] for info in data_dict['response']['body']['item']['applicantInfoArray']['applicantInfo']]
            item['applicantName'] = ', '.join(names_list)
        else:
            item['applicantName'] = data_dict['response']['body']['item']['applicantInfoArray']['applicantInfo']['name']
        item['applicationDate'] = data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('applicationDate')
        item['applicationNumber'] = no
        item['astrtCont'] = data_dict['response']['body']['item']['abstractInfoArray']['abstractInfo'].get('astrtCont')
        item['bigDrawing'] = data_dict['response']['body']['item']['imagePathInfo'].get('largePath')
        item['drawing'] = data_dict['response']['body']['item']['imagePathInfo'].get('path')
        item['indexNo'] = "1"
        item['inventionTitle'] = data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('inventionTitle')
        ipc_info = data_dict['response']['body']['item']['ipcInfoArray']['ipcInfo']
        if isinstance(ipc_info, list):
            item['ipcNumber'] = '|'.join([i.get('ipcNumber', '') for i in ipc_info])
        else:
            item['ipcNumber'] = ipc_info.get('ipcNumber', '')
        item['openDate'] = safe_replace(data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('openDate'), '.', '')
        item['openNumber'] = safe_replace(data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('openNumber'), '-', '')
        item['publicationDate'] = safe_replace(data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('publicationDate'), '.', '')
        item['publicationNumber'] = safe_replace(data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('publicationNumber'), '-', '')
        item['registerDate'] = safe_replace(data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('registerDate'), '.', '')
        item['registerNumber'] = safe_replace(data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('registerNumber'), '-', '')
        item['registerStatus'] = data_dict['response']['body']['item']['biblioSummaryInfoArray']['biblioSummaryInfo'].get('registerStatus')
        item['title'] = data['data'][i]['title']
        item['entire_text'] = data['data'][i]['summary_entire'][0]['orginal_text']
        item['section_text'] = data['data'][i]['summary_section'][0]['orginal_text']
        return item
    else :
        return '없음'
```

```python
def run(api_list):
    base_url_list = [
        'http://kipo-api.kipi.or.kr/openapi/service/patUtiModInfoSearchSevice/getWordSearch',
        'http://kipo-api.kipi.or.kr/openapi/service/patUtiModInfoSearchSevice/getAdvancedSearch',
        'http://kipo-api.kipi.or.kr/openapi/service/patUtiModInfoSearchSevice/getBibliographyDetailInfoSearch'
    ]

    all_items = []
    all_null = []
    # 파일 경로
    file_path1 = 'full_text_data.json'
    with open(file_path1, 'r', encoding='utf-8') as json_file:
        data = json.load(json_file)

    file_path2 = 'full_data.json'
    # 파일 존재 여부 확인
    if os.path.exists(file_path2):
        # 파일 읽기
        with open(file_path2, 'r', encoding='utf-8') as json_file:
            done_data = json.load(json_file)
        
        # 데이터에서 필요한 값 추출
        origin_data = done_data.get('items', [])
        
        # 기존 데이터에 새로운 데이터를 추가
        all_items.extend(origin_data)
        print(f"수집된 데이터가 {len(origin_data)}개 존재합니다.")
    else:
        print("파일이 존재하지 않습니다.")

    file_path3 = 'null_data.json'
    # 파일 존재 여부 확인
    if os.path.exists(file_path3):
        # 파일 읽기
        with open(file_path3, 'r', encoding='utf-8') as json_file:
            null_data = json.load(json_file)
        
        # 데이터에서 필요한 값 추출
        null_data = null_data.get('items', [])
        
        # 기존 데이터에 새로운 데이터를 추가
        all_null.extend(null_data)
        print(len(null_data), "개의 null 데이터가 존재합니다.")
    else:
        print("null_data.json이 존재하지 않습니다.")

    num = len(all_items) + len(all_null)
    check = 1
    for api_key in api_list:
        print(f'{len(api_list)}개의 API 중 {check}번째 실행')
        check += 1
        params_list = [
            {'serviceKey': api_key,
             'year': '0',
             'patent': 'true',
             'utility': 'true',
             'numOfRows': '1000'},
            {'serviceKey': api_key,
             'year': '0',
             'patent': 'true',
             'utility': 'true',
             'numOfRows': '1000'},
            {'serviceKey': api_key}
        ]
        
        for k in range(1, len(base_url_list)):
            base_url = base_url_list[k]
            params = params_list[k]

            initial_items_len = len(all_items)
            initial_null_len = len(all_null)

            for i in range(num, len(data['data'])):
                no = data['data'][i]['reg_no']
                if k == 0:
                    params['word'] = no
                else:
                    params['applicationNumber'] = no
                response = requests.get(base_url, params=params)
                data_dict = xmltodict.parse(response.content)
                if data_dict['response']['header']['resultCode'] != '00':
                    print(data_dict['response']['header']['resultMsg'])
                    num = i
                    break
                else:                    
                    try:
                        if k == 0:
                            result = url1(data, i, no, data_dict)
                            if result != '없음':
                                all_items.append(result)
                            else:
                                null_dict = {'applicationNumber': no}
                                all_null.append(null_dict)
                        elif k == 1:
                            result = url2(data, i, data_dict)
                            if result != '없음':
                                all_items.append(result)
                            else:
                                null_dict = {'applicationNumber': no}
                                all_null.append(null_dict)
                        else:
                            result = url3(data, i, no, data_dict)
                            if result != '없음':
                                all_items.append(result)
                            else:
                                null_dict = {'applicationNumber': no}
                                all_null.append(null_dict)
                    except Exception as e:
                        print(f"오류 발생 {no}: {e}")
                        null_dict = {'applicationNumber': no}
                        all_null.append(null_dict)

            # 변경된 데이터가 있는 경우에만 파일 저장
            if len(all_items) != initial_items_len:
                final_data = {'items': all_items}
                json_data = json.dumps(final_data, indent=4, ensure_ascii=False)

                with open('full_data.json', 'w', encoding='utf-8') as json_file:
                    json_file.write(json_data)
                    print('API 실행 후 수집된 데이터 : ', len(all_items))
                    print('저장 완료')
            else:
                print('새로운 데이터 추가 없음으로 넘어감')
                
            if len(all_null) != initial_null_len:
                final_null_data = {'items': all_null}
                json_null_data = json.dumps(final_null_data, indent=4, ensure_ascii=False)

                with open('null_data.json', 'w', encoding='utf-8') as json_file:
                    json_file.write(json_null_data)
```

```python
import json
import requests
import time

def get_pdf_url_for_missing_data(key_list):
    file_path = '/media/user/datadisk/Patent_data/full_data.json'
    
    # 파일 읽기
    with open(file_path, 'r', encoding='utf-8') as json_file:
        done_data = json.load(json_file)

    data = done_data.get('items', [])
    url_list = {
        'pdf_open': 'http://kipo-api.kipi.or.kr/openapi/service/patUtiModInfoSearchSevice/getPubFullTextInfoSearch',
        'pdf_register': 'http://kipo-api.kipi.or.kr/openapi/service/patUtiModInfoSearchSevice/getAnnFullTextInfoSearch',
        'pdf_rewrite_register': 'http://kipo-api.kipi.or.kr/openapi/service/patUtiModInfoSearchSevice/getRevisionFullTextInfoSearch'
    }

    key_num = 0
    print(f'{key_num + 1}번째 API 시작')

    updated_items = 0  # 업데이트된 항목 수 추적

    # 키가 없는 데이터 필터링
    missing_pdf_keys = [item for item in data if 'pdf_open' not in item and 'pdf_register' not in item and 'pdf_rewrite_register' not in item]
    print(len(missing_pdf_keys))
    
    # 키가 없는 데이터들에 대해 URL 추가
    for i, item in enumerate(missing_pdf_keys):
        updated = False  # 데이터가 업데이트되었는지 여부를 확인하기 위한 플래그

        for url_name, url in url_list.items():
            # API 호출을 통해 URL 추가
            params = {'serviceKey': key_list[key_num], 'applicationNumber': item['applicationNumber']}
            
            for attempt in range(3):  # 최대 3번 재시도
                try:
                    response = requests.get(url, params=params, timeout=10)
                    result = response.content.decode('utf-8')
                    print(f"전체: {result}")
                    if 'LIMITED NUMBER' in result:
                        key_num += 1
                        if key_num >= len(key_list):
                            save_data_to_file(data, file_path)
                            print('API 키 소진, 작업 종료.')
                            return
                        else:
                            print(f'{key_num + 1}번째 API 키로 전환')
                        break  # 다른 키로 이동하기 위해 반복문 탈출

                    # 성공 시 URL을 item에 추가
                    print(f"전 result: {result}")
                    if 'path>' in result:
                        print(f"후 result: {result}")
                        item[url_name] = result.split('path>')[1].split('<')[0]
                    else:
                        item[url_name] = 'null'

                    updated = True  # 데이터 업데이트 상태 설정
                    print(f"{i + 1}번째 항목에 {url_name} URL 추가 완료")
                    updated_items += 1
                    break  # 성공적으로 처리되었으므로 반복문 탈출

                except requests.exceptions.Timeout:
                    print(f"요청 시간 초과, 재시도 중... ({attempt + 1}/3)")
                    time.sleep(2)  # 2초 대기 후 재시도
            else:
                print("재시도 횟수 초과, 다음 항목으로 이동")
                item[url_name] = 'null'
                updated = True  # 'null'로 처리한 경우에도 업데이트 상태 설정

        # 업데이트된 항목만 저장
        if updated:
            print(f"{i + 1}번째 항목 업데이트 완료")

    # 모든 항목 업데이트 후 파일 저장
    save_data_to_file(data, file_path)
    print(f'총 {updated_items}개의 항목이 업데이트되었습니다.')
    print('끝~~~')

def save_data_to_file(data, file_path):
    final_data = {'items': data}
    json_data = json.dumps(final_data, indent=4, ensure_ascii=False)
    with open(file_path, 'w', encoding='utf-8') as json_file:
        json_file.write(json_data)
        print('저장 완료')

get_pdf_url_for_missing_data(api_list)

# 이후 JSON 파일을 다시 읽고 키가 없는 데이터 개수를 확인하는 코드:
file_path = '/media/user/datadisk/Patent_data/full_data.json'

with open(file_path, 'r', encoding='utf-8') as json_file:
    done_data = json.load(json_file)

data = done_data.get('items', [])
missing_pdf_keys = [item for item in data if 'pdf_open' not in item and 'pdf_register' not in item and 'pdf_rewrite_register' not in item]

print(f"키가 없는 데이터 개수: {len(missing_pdf_keys)}")
print(f"총 데이터 수: {len(data)}")
```

```python
def save_pdf(save_folder, filename, url):
    filename += '.pdf'
    file_path = os.path.join(save_folder, filename)
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    file.write(chunk)
        print(f"PDF 파일이 {file_path}에 성공적으로 저장되었습니다.", end=' ')
        return file_path
    else:
        print("PDF 파일 다운로드에 실패했습니다.")
        return None    
```

```python
def extract_and_split_first_page_text(first_page):
    """
    첫 번째 페이지에서 '등록일자' 다음 줄부터 '전체 청구항 수' 전 줄까지의 텍스트를
    세로로 나누어 각 부분의 텍스트를 반환하고,
    나머지 텍스트는 여백(왼쪽 10%, 위쪽 5%, 아래쪽 5%)을 제외하고 추출합니다.
    """
    words = first_page.extract_words()

    # 단어들을 Y좌표를 기준으로 라인별로 그룹화
    def group_words_into_lines(words, y_tolerance=2):
        lines = []
        words_sorted = sorted(words, key=lambda w: w['top'])
        current_line = []
        current_top = None

        for word in words_sorted:
            if current_top is None:
                current_top = word['top']
                current_line.append(word)
            else:
                if abs(word['top'] - current_top) <= y_tolerance:
                    current_line.append(word)
                else:
                    lines.append(current_line)
                    current_line = [word]
                    current_top = word['top']
        if current_line:
            lines.append(current_line)
        return lines

    lines = group_words_into_lines(words)

    # 각 라인의 텍스트와 위치 정보 추출
    line_texts = []
    for line in lines:
        text = ' '.join([w['text'] for w in line])
        top = min([w['top'] for w in line])
        bottom = max([w['bottom'] for w in line])
        left = min([w['x0'] for w in line])
        right = max([w['x1'] for w in line])
        line_texts.append({'text': text, 'top': top, 'bottom': bottom, 'left': left, 'right': right})

    # '등록일자'와 '전체 청구항 수' 라인의 인덱스 찾기
    start_line_index = None
    end_line_index = None

    for idx, line in enumerate(line_texts):
        if '(24) 등록일자' in line['text']:
            start_line_index = idx + 1  # '등록일자' 다음 줄부터 시작
        if '전체 청구항 수' in line['text']:
            end_line_index = idx - 1  # '전체 청구항 수' 전 줄까지
            break  # 첫 번째로 발견된 '전체 청구항 수'를 기준으로 함

    extracted_bbox = None
    left_half_text = right_half_text = None

    if start_line_index is not None and end_line_index is not None and start_line_index <= end_line_index:
        # 관심 영역의 좌표 계산
        top_positions = [line_texts[i]['top'] for i in range(start_line_index, end_line_index+1)]
        bottom_positions = [line_texts[i]['bottom'] for i in range(start_line_index, end_line_index+1)]
        left_positions = [line_texts[i]['left'] for i in range(start_line_index, end_line_index+1)]
        right_positions = [line_texts[i]['right'] for i in range(start_line_index, end_line_index+1)]

        bbox_left = min(left_positions)
        bbox_top = min(top_positions)
        bbox_right = max(right_positions)
        bbox_bottom = max(bottom_positions)

        extracted_bbox = (bbox_left, bbox_top, bbox_right, bbox_bottom)

        # 영역을 세로로 반으로 나누기
        mid_x = bbox_left + (bbox_right - bbox_left) / 2

        left_half_bbox = (bbox_left, bbox_top, mid_x, bbox_bottom)
        right_half_bbox = (mid_x, bbox_top, bbox_right, bbox_bottom)

        # 각 영역에서 텍스트 추출
        left_half_text = first_page.crop(bbox=left_half_bbox).extract_text()
        right_half_text = first_page.crop(bbox=right_half_bbox).extract_text()
    else:
        extracted_bbox = None
        left_half_text = "해당 키워드를 찾을 수 없습니다."
        right_half_text = None

    # 첫 페이지의 나머지 텍스트 추출 (여백 제외하고, 이미 추출된 영역 제외)
    page_width = first_page.width
    page_height = first_page.height

    # 여백을 제외한 전체 영역
    margin_left = page_width * 0.1
    margin_top = page_height * 0.05
    margin_right = page_width
    margin_bottom = page_height * 0.95

    # 여백을 제외한 영역에서 단어 추출
    words_in_main_area = [
        word for word in words
        if (margin_left <= (word['x0'] + word['x1']) / 2 <= margin_right) and
           (margin_top <= (word['top'] + word['bottom']) / 2 <= margin_bottom)
    ]

    # 이미 추출된 영역의 단어를 제외
    if extracted_bbox:
        words_in_rest_area = [
            word for word in words_in_main_area
            if not (
                extracted_bbox[0] <= (word['x0'] + word['x1']) / 2 <= extracted_bbox[2] and
                extracted_bbox[1] <= (word['top'] + word['bottom']) / 2 <= extracted_bbox[3]
            )
        ]
    else:
        words_in_rest_area = words_in_main_area

    # 나머지 텍스트 추출
    rest_text = ' '.join([word['text'] for word in words_in_rest_area])

    return left_half_text.strip(), right_half_text.strip(), rest_text.strip()

def extract_main_text_from_other_pages(page):
    """
    첫 페이지 이외의 페이지에서 본문 텍스트만 추출.
    페이지의 상단, 하단, 왼쪽 영역을 제외하고 본문만 추출.
    """
    height = page.height
    width = page.width

    # 페이지의 전체 텍스트를 위치 정보와 함께 추출
    words = page.extract_words()

    # 본문 텍스트를 저장할 리스트
    body_text = []

    for word in words:
        # 각 텍스트의 y 좌표 (세로 위치)와 x 좌표 (가로 위치) 확인
        top_position = (word['top'] + word['bottom']) / 2
        left_position = (word['x0'] + word['x1']) / 2

        # 상단 5%, 하단 5%, 왼쪽 10% 영역 제외
        if height * 0.05 < top_position < height * 0.95 and left_position > width * 0.1:
            body_text.append(word['text'])

    return ' '.join(body_text).strip()

def extract_text_from_pdf(pdf_path):
    """
    PDF에서 첫 페이지는 지정된 영역을 세로로 나누어 추출하고,
    나머지 텍스트는 여백을 제외하고 추출합니다.
    나머지 페이지는 본문만 추출합니다.
    """
    with pdfplumber.open(pdf_path) as pdf:
        full_text = ''

        # 첫 번째 페이지 처리
        first_page = pdf.pages[0]
        left_text, right_text, rest_text = extract_and_split_first_page_text(first_page)
        
        split_text = rest_text.split('전체 청구항 수')
        full_text+= split_text[0] + left_text.replace('\n',' ') + right_text.replace('\n',' ')+ ' 전체 청구항 수' +split_text[1]
        # 나머지 페이지 처리
        for page_num in range(1, len(pdf.pages)):
            page = pdf.pages[page_num]
            page_text = extract_main_text_from_other_pages(page)
            full_text += page_text

    return full_text
```
